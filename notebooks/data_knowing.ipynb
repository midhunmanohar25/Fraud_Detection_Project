# Sampling the dataset


```python
orginal_df.head()
```

```python
sample_df = orginal_df.sample(frac=0.2, random_state=42)
```

```python
sample_df.head()
```

```python
sample_df.shape
```

```python
sample_df.drop(columns=['transaction_id', 'customer_id', 'device_id', 'merchant_id', 'timestamp'], inplace=True)
```

```python
from sklearn.preprocessing import OrdinalEncoder

oe = OrdinalEncoder()

categorical_col = sample_df.select_dtypes('object')
for col in categorical_col:
    sample_df[col] = oe.fit_transform(sample_df[[col]])
```

```python
sample_df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))
sns.heatmap(sample_df.corr())
```
